collaborative filtering

```python
rating_df.head()
```

```python
rating_df.groupby('User-ID').count()['Book-Rating']
```

```python
reliable_users = rating_df.groupby('User-ID').count()['Book-Rating'] > 200
reliable_users = reliable_users[reliable_users].index
reliable_users.shape
```

```python
rating_df = rating_df[rating_df['User-ID'].isin(reliable_users)]
rating_df.shape
```

```python
ra = rating_df.groupby(['Book-Title']).count()['Book-Rating'] > 50
famous_books = ra[ra].index
famous_books
```

```python
famous_books.shape
```

```python
 final_rating = rating_df[rating_df['Book-Title'].isin(famous_books)]
 final_rating.head()
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
pt.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
similarity_score.shape
```

```python
def recommend(book_name):
  index = np.where(pt.index==book_name)[0][0]
  similar = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1],reverse=True)[1:6]
  for i in similar:
    print(pt.index[i[0]])
```

```python
recommend('1984')
```
